```python
import topojson as tp
import geopandas as gpd

world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
africa = world.query('continent == "Africa"')
```

```python
l = tp.Topology(data=africa, topology=False).toposimplify(4).to_alt().properties(title='WITHOUT Topology')
r = tp.Topology(africa, topology=True).toposimplify(4).to_alt().properties(title='WITH Topology')
l | r
```

```python
l = tp.Topology(africa).to_alt().properties(title='NO Topoquantization')
r = tp.Topology(africa).topoquantize(100).to_alt().properties(title='WITH Topoquantization')
l | r
```

```python
topo = tp.Topology(africa) 

topo.options.simplify_with = 'shapely'
topo.options.simplify_algorithm = 'dp' 
l = topo.toposimplify(4).to_alt().properties(title=['Douglas-Peucker simplification'])

topo.options.simplify_with = 'simplification'
topo.options.simplify_algorithm = 'vw' 
r = topo.toposimplify(4).to_alt().properties(title='Visvalingam-Whyatt simplification')
l | r
```

```python
africa.to_file('africa.geo.json', driver='GeoJSON')
```

```python
topo = tp.Topology(africa, simplify_with='simplification', simplify_algorithm='vw', toposimplify=4, topoquantize=200)
topo.to_json('africa.topo.json')
```

```python
import os
size_geojson = os.stat('africa.geo.json').st_size
size_topojson = os.stat('africa.topo.json').st_size

print(f'africa.geo.json: {size_geojson/1000} KB')
print(f'africa.topo.json: {size_topojson/1000} KB')
print(f'factor smaller: {size_geojson/size_topojson} X')
```

```python
l = tp.Topology(data=africa, topology=False, prequantize=False).to_alt().properties(title=['Raw GeoJSON', '102.72 KB'])
r = topo.to_alt().properties(title=['Optimized TopoJSON', '13.21 KB (7.8X smaller)'])
l | r
```
